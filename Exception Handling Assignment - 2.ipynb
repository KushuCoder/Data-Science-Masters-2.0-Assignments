# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.


```python
#The LookupError class is the base class for errors related to invalid key/index lookups in sequences or mappings. It helps catch errors when accessing non-existent elements.
#key error
try:
    data = {"name": "Alice"}
    print(data["age"]) 
except KeyError as e:
    print(f"Error: Key {e} not found!")
#Index error
try:
    numbers = [1, 2, 3]
    print(numbers[5])  
except IndexError as e:
    print(f"Error: {e}")
```

# Q5. Explain ImportError. What is ModuleNotFoundError?


ImportError raises when an import statement fails due to a missing module or incorrect import.      
ModuleNotFoundError is a subclass of ImportError, specifically for missing module cases.

# Q6. List down some best practices for exception handling in python.

-Avoid Using raise Unnecessarily, Only raise exceptions when necessary, and use meaningful messages.        
-Use else for Code That Should Run Only if No Exception Occurs, Keeps the try block clean and avoids unintended execution.         
-Create Meaningful Custom Exceptions, Inherit from Exception and use descriptive error messages.       
-Use Built-in Exceptions Where Possible, Avoid unnecessary custom exceptions if a built-in one suffices.
